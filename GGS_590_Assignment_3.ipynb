# Task 1
Create three Shapely objects for a point, a linestring and a polygon. Add these to a list, and then
iterate over the contents. Print the bounds and area of each geometry object. Explain the results you
obtain for the different geometry types (using at least 30 words of explanation per type).

```python
from shapely.geometry import Point, LineString, Polygon

# Create Shapely objects
my_point = Point(5,10)
my_linestring = LineString([(2,3), (8,6), (12,9)])
my_polygon = Polygon([(4,4),(6,8),(9,5),(4,4)])

# Add these objects to a list
geometry_list = [my_point, my_linestring, my_polygon]

# Iterate over the geometry_list and print bounds and area
for geom in geometry_list:
    print(f"Geometry: {geom}")
    print(f"Bounds: {geom.bounds}")
    print(f"Area: {geom.area}\n")
    print()

# Explanation
print("Explanation of the results:")
print("1. The my_point has no area, as it represents a single location in space. Its bounds are the coordinates of the point itself, which are (5.0, 10.0).")
print("2. The my_linestring is a a one-dimensional object with length but no area. Its bounds are defined by the minimum and maximum coordinates of the points that make up the linestring, which are (2.0, 3.0) and (12.0, 9.0).")
print("3. The my_polygon is a two-dimensional object that encloses a space. Its bounds are are the minimum and maximum coordinates of its vertices, which are (4.0, 4.0) and (9.0, 8.0). The area is the space enclosed by the polygon, which is 9.0 square units.")
```

# Task # 2
Iterate over your list of geometry objects and check the geometry type. Using loop control logic, print the bounds of any points, print the length of any linestring, and print the area of any polygon.

```python
from shapely.geometry import Point, LineString, Polygon

# List of geometry objects
geometries = [
    Point(3, 6),
    LineString([(2, 2), (3, 3), (4, 4)]),
    Polygon([(6, 5), (1, 1), (0, 0),(6, 5)])
]

# Iterate over the geometry objects
for geom in geometries:
    if geom.geom_type == "Point":
        print("Point bounds:", geom.bounds)
    elif geom.geom_type == "LineString":
        print("LineString length:", geom.length)
    elif geom.geom_type == "Polygon":
        print("Polygon area:", geom.area)
```

# Task # 5
Write a program based on a real-world problem for a set of geometries demonstrating the use of at least three geometric operations (e.g., union, intersection, buffer etc.). It would be wise to try apply these as a demonstrator for a topic of interest that you might utilize in your final coursework project.

```python
from shapely.geometry import LineString, Polygon
from shapely.ops import unary_union
import matplotlib.pyplot as plt

# Roads as LineStrings, it representing roads in a traffic network
road_1 = LineString([(0, 0), (4, 4)])
road_2 = LineString([(1, 0), (5, 4)])
road_3 = LineString([(0, 4), (4, 0)])

# Geometric operations
# 1. buffer: The buffer around each road represent traffic flow zones (e.g., impact area of the road)
buffer1 = road_1.buffer(0.5)
buffer2 = road_2.buffer(0.5)
buffer3 = road_3.buffer(0.5)

# 2. Union: By combining the buffers into a single geometry it can analyze the combined effect of traffic zones
traffic_zone_union = unary_union([buffer1, buffer2, buffer3])

# 3. Intersection: The intersections between different road is buffer to determine congestion areas
congestion_area = buffer1.intersection(buffer2)

# Plot the geometries
fig, ax = plt.subplots()

# Plot roads
x1, y1 = road_1.xy
x2, y2 = road_2.xy
x3, y3 = road_3.xy
ax.plot(x1, y1, color='blue', label='Road 1')
ax.plot(x2, y2, color='green', label='Road 2')
ax.plot(x3, y3, color='red', label='Road 3')

# Plot buffers (traffic zones)
x, y = buffer1.exterior.xy
ax.fill(x, y, alpha=0.4, color='blue', label='Traffic Zone 1')

x, y = buffer2.exterior.xy
ax.fill(x, y, alpha=0.4, color='green', label='Traffic Zone 2')

x, y = buffer3.exterior.xy
ax.fill(x, y, alpha=0.4, color='red', label='Traffic Zone 3')

# Plot union of traffic zones
x, y = traffic_zone_union.exterior.xy
ax.plot(x, y, color='black', linestyle='--', label='Union of Zones')

# Plot congestion area (intersection of two buffers)
if not congestion_area.is_empty:
    x, y = congestion_area.exterior.xy
    ax.fill(x, y, alpha=0.6, color='purple', label='Congestion Area')

ax.legend()
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Traffic Flow Analysis')
plt.show()
```
